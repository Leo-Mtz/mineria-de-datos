#### Conceptos

La minería de datos, en términos sencillos, es el proceso de descubrir patrones, tendencias y conocimientos útiles a partir de grandes conjuntos de datos. Es como encontrar una aguja en un pajar, pero con información. Se utiliza para predecir comportamientos, identificar riesgos, mejorar decisiones y mucho más.

##### Ciencia de Datos

La ciencia de datos es el campo que usa métodos científicos, algoritmos y sistemas para extraer conocimiento y 'insights' de los datos, combinando estadísticas, programación y conocimiento del negocio.
